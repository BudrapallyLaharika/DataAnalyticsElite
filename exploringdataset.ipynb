```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
diamonds = sns.load_dataset('diamonds')
print(diamonds.head())
```

```python
# Checking for missing values
print(diamonds.isnull().sum())
# Getting dataset info
print(diamonds.info())
# Summary of statistics
print(diamonds.describe())
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(diamonds['carat'], kde=True)
plt.title('Distribution of Carat')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='cut', y='price', data=diamonds)
plt.title('Price Distribution by Cut Quality')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='color', y='price', data=diamonds)
plt.title('Price Distribution by Diamond Color')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='carat', y='price', data=diamonds, hue='cut', alpha=0.6)
plt.title('Carat vs. Price')
plt.show()
```
